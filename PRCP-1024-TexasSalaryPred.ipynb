## DOMAIN ANALYSIS

**1.AGENCY**: A numeric identifier for the government agency or organization employing the individual.

**2.AGENCY NAME**: The name of the government agency or organization.

**3.LAST NAME**: The employee's last name.

**4.FIRST NAME**: The employee's first name.

**5.MI**: The employee's middle initial (if available).

**6.CLASS CODE**: A code representing the employee’s job classification.

**7.CLASS TITLE**: The title or designation of the employee’s job.

**8.ETHNICITY**: The ethnicity of the employee (e.g., White, Hispanic, etc.).

**9.GENDER**: The gender of the employee (e.g., Male, Female).

**10.STATUS**: The employment status of the employee, indicating whether they are full-time, part-time, temporary, or regular.

**11.EMPLOY DATE**: The date when the employee started working for the organization.

**12.HRLY RATE**: The employee’s hourly wage rate.

**13.HRS PER WK**: The number of hours the employee works per week.

**14.MONTHLY**: The employee’s calculated monthly salary.

**15.ANNUAL**: The employee’s calculated annual salary.

**16.STATE NUMBER**: Another numeric identifier, potentially related to the employee’s position or payroll record.

**17.duplicated**: Indicates whether this entry is a duplicate record (e.g., "True" or "NaN").

**18.multiple_full_time_jobs**: Flags employees holding multiple full-time jobs (e.g., 1 for true, NaN for false).

**19.combined_multiple_jobs**: Describes combined information about multiple jobs if applicable.

**20.summed_annual_salary**: The total annual salary when combining multiple job records for an individual.

**21.hide_from_search**: Indicates whether this record is hidden from public searches (e.g., "True" or "NaN").

#### Upper bound And Lower bound

```python
q3  = np.percentile(data.ANNUAL_SALARY,75)
q1  = np.percentile(data.ANNUAL_SALARY, 25)

IQR = q3 - q1
upperbound = q3 + 1.5*IQR
lowerbound = q1 - 1.5*IQR

print(f"LOWERBOUND :{lowerbound}")
print(f"UPPERBOUND :{upperbound}")
```

```python
data.loc[(data.ANNUAL_SALARY>upperbound) | (data.ANNUAL_SALARY<lowerbound)].shape# these are the annual salary outliers
```

 By utilizing box plots and establishing lower and upper limits, we initially identified 9030 outliers. However, upon closer inspection through domain analysis, we discerned that many of these outliers may not truly be outliers. Our refined definition of outliers focuses on instances where there are substantial gaps between data points, and they are significantly positioned away from both the upper and lower limits. This nuanced approach allows us to more accurately identify and consider outliers within the dataset.

### Model Creation

#### LinearRegression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
# Linear Regression
lr = LinearRegression()
lr.fit(X_train_scaled,y_train_scaled)
y_test_pred_lr =  lr.predict(X_test_scaled)# prediction for testing data
y_train_pred_lr =  lr.predict(X_train_scaled)# prediction for training data
```

```python
# Evaluate on the training set
print("Training Set Evaluation Linear Regression:")
print("MAE:", mean_absolute_error(y_true=y_train_scaled, y_pred=y_train_pred_lr))
print("MSE:", mean_squared_error(y_train_scaled, y_train_pred_lr))
print("R² Score:", r2_score(y_train_scaled, y_train_pred_lr))

# Evaluate on the test set
print("\nTest Set Evaluation Linear Regression:")
print("MAE:", mean_absolute_error(y_test_scaled, y_test_pred_lr))
print("MSE:", mean_squared_error(y_test_scaled, y_test_pred_lr))
print("R² Score:", r2_score(y_true=y_test_scaled,y_pred= y_test_pred_lr))
```

#### DecisionTreeRegressor

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

```python
# decision tree regressor
dtr =  DecisionTreeRegressor(max_depth= 25,min_samples_split=10,random_state=9)
dtr.fit(X_train,y_train)# training with unscaled data, tree based models do not require scaled data
y_test_pred_dtr =  dtr.predict(X_test)
y_train_pred_dtr =  dtr.predict(X_train)
```

```python
# Evaluate on the training set
print("Training Set Evaluation Descion Tree Regressor:")
print("MAE:", mean_absolute_error(y_train,y_train_pred_dtr))
print("MSE:", mean_squared_error(y_train,y_train_pred_dtr))
print("R² Score:", r2_score(y_train,y_train_pred_dtr))

# Evaluate on the test set
print("\nTest Set Evaluation Descion Tree Regressor:")
print("MAE:", mean_absolute_error(y_test,y_test_pred_dtr))
print("MSE:", mean_squared_error(y_test,y_test_pred_dtr))
print("R² Score:", r2_score(y_test,y_test_pred_dtr))
```

```python
# Define the parameter grid for DecisionTreeRegressor
param_grid = {
    'max_depth': [3, 5, 10, None],            # Tree depth
    'min_samples_split': [2, 5, 10],          # Minimum samples to split a node
    'min_samples_leaf': [1, 2, 5],            # Minimum samples per leaf
    'max_features': [None, 'sqrt', 'log2'],   # Features to consider for split
    'criterion': ['squared_error', 'friedman_mse']  # Loss function
}
```

```python
### Decision tree with grid search cv
```

```python
tree = DecisionTreeRegressor(random_state=42)

grid_search = GridSearchCV(estimator=tree, param_grid=param_grid,
                           cv=5, scoring='neg_mean_squared_error', verbose=1)

grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)

best_tree = grid_search.best_estimator_

y_train_pred_grid = best_tree.predict(X_train)
y_test_pred_grid = best_tree.predict(X_test)

# Training Metrics
train_mae = mean_squared_error(y_train, y_train_pred_grid)
print("Training MSE:\n", train_mae)

# Testing Metrics
test_mae = mean_squared_error(y_test, y_test_pred_grid)
print("Testing MSE:\n", test_mae)


# Evaluate on the training set
print("\nTraining Set Evaluation Descion Tree Regressor grid:")
print("MAE:", mean_absolute_error(y_train,y_train_pred_grid))
print("MSE:", mean_squared_error(y_train,y_train_pred_grid))
print("R² Score:", r2_score(y_train,y_train_pred_grid))

# Evaluate on the test set
print("\nTest Set Evaluation Descion Tree Regressor grid :")
print("MAE:", mean_absolute_error(y_test,y_test_pred_grid))
print("MSE:", mean_squared_error(y_test,y_test_pred_grid))
print("R² Score:", r2_score(y_test, y_test_pred_grid))
```
